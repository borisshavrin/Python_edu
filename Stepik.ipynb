Чтобы быстро вычислять жилплощадь, требуется написать программу, на вход которой подаётся тип фигуры комнаты и соответствующие параметры, которая бы выводила площадь получившейся комнаты.
Для числа π в стране Малевии используют значение 3.14.

Формат ввода, который используют Малевийцы:

треугольник
a
b
c
где a, b и c — длины сторон треугольника

прямоугольник
a
b
где a и b — длины сторон прямоугольника

круг
r
где r — радиус окружности

```python
type = input() # программа просит пользователя ввести тип площади его комнаты (треугольник, прямоугольник или круг)
pi = 3.14      # обозначили число Пи

if type == 'треугольник':   # если пользователь ввел тип 'треугольник', то программа просит ввести значения длин сторон тр-ка
    a = int(input())
    b = int(input())
    c = int(input())
    p = (a + b + c) / 2     # формула полупериода треугольника
    S_triangle = (p * (p - a) * (p - b) * (p -c)) ** (1 / 2) #формула Герона для вычисления площади треугольника по 3м сторонам
    
    print(S_triangle)
    
if type == 'прямоугольник': # если пользователь ввел тип 'прямоугольник', то программа просит ввести значения длин сторон пр-ка
    a = int(input())
    b = int(input())
    S_rectangle = a * b     # формула площади прямоугольника
    
    print(S_rectangle)
    
if type == 'круг':          # если пользователь ввел тип 'круг', то программа просит ввести значения длины радиуса круга
    r = int(input())
    S_circle = pi * (r ** 2)   # формула площади круга
    
    print(S_circle)
```

Задача 2.3.7: Напишите программу, которая считывает с клавиатуры два числа a и b, считает и выводит на консоль среднее арифметическое всех чисел из отрезка [a; b], которые делятся на 3.

В приведенном ниже примере среднее арифметическое считается для чисел на отрезке [-5; 12]. Всего чисел, делящихся на 3, на этом отрезке 6: -3, 0, 3, 6, 9, 12. Их среднее арифметическое равно 4.5

На вход программе подаются интервалы, внутри которых всегда есть хотя бы одно число, которое делится на 3.

```python
a = int(input())           # ввод границ интервала
b = int(input())

s = []                     # создание накопителя чисел
for i in range(a, b + 1):  # каждое число в интервале от а до b,
    if i % 3 == 0:         # остаток которого при делении на 3 равен 0,
        s.append(i)        # добавляется в накопитель s

sum_all = 0                # накопитель суммы
for j in s:                # к каждому числу из накопителя s прибавляется следующее число
    sum_all += j

x = sum_all / len(s)       # среднее арифметическое = накопитель суммы / количество чисел в накопителе s
print(x)
```

Для элементов списка, являющихся крайними, одним из соседей считается элемент, находящий на противоположном конце этого списка. Например, если на вход подаётся список "1 3 5 6 10", то на выход ожидается список "13 6 9 15 7" (без кавычек).

Если на вход пришло только одно число, надо вывести его же.

Вывод должен содержать одну строку с числами нового списка, разделёнными пробелом.

```python
list_of_numbers = [int(i) for i in input().split()]  # ввод списка чисел одной строкой

if len(list_of_numbers) == 1:     # если список состоит из одного числа, то выводится это самое число
    print(list_of_numbers[0])

else:
    new_list = []    # создание нового списка, куда будут записываться суммы 
    x = 0
    for i in list_of_numbers:
        while x <= len(list_of_numbers) - 2:
            number_left = list_of_numbers[x - 1]
            number_right = list_of_numbers[x + 1]
            z = number_left + number_right
            new_list.append(z)
            x += 1
        if x > len(list_of_numbers) - 2:
            number_left = list_of_numbers[x - 1]
            number_right = list_of_numbers[0]
            z = number_left + number_right
            new_list.append(z)
            break
    for i in new_list:
        print(i, end=' ')
```

Напишите программу, которая принимает на стандартный вход список игр футбольных команд с результатом матча и выводит на стандартный вывод сводную таблицу результатов всех матчей.
За победу команде начисляется 3 очка, за поражение — 0, за ничью — 1.

Формат ввода следующий:
В первой строке указано целое число nn — количество завершенных игр.
После этого идет nn строк, в которых записаны результаты игры в следующем формате:
Первая_команда;Забито_первой_командой;Вторая_команда;Забито_второй_командой

Вывод программы необходимо оформить следующим образом:
Команда:Всего_игр Побед Ничьих Поражений Всего_очков

Конкретный пример ввода-вывода приведён ниже.

Порядок вывода команд произвольный.

Sample Input:

3

Спартак;9;Зенит;10

Локомотив;12;Зенит;3

Спартак;8;Локомотив;15


Sample Output:

Спартак:2 0 0 2 0
Зенит:2 1 0 1 3
Локомотив:2 2 0 0 6

```python
total_games = int(input())
input_table = [input().split(';') for i in range(total_games)]

numbers = set(str(i) for i in range(100))

n_games_team = {team: 0 for row in input_table for team in row if team not in numbers}
n_win_team = {team: 0 for row in input_table for team in row if team not in numbers}
n_draw_team = {team: 0 for row in input_table for team in row if team not in numbers}
n_lose_team = {team: 0 for row in input_table for team in row if team not in numbers}
all_points = {team: 0 for row in input_table for team in row if team not in numbers}

for row in input_table:
    for team in row:
        if team in n_games_team:
            n_games_team[team] += 1

for row in input_table:
    if int(row[1]) > int(row[3]):
        n_win_team[row[0]] += 1
        n_lose_team[row[2]] += 1
        all_points[row[0]] += 3
    if int(row[1]) < int(row[3]):
        n_win_team[row[2]] += 1
        n_lose_team[row[0]] += 1
        all_points[row[2]] += 3
    if int(row[1]) == int(row[3]):
        n_draw_team[row[0]] += 1
        n_draw_team[row[2]] += 1
        all_points[row[0]] += 1
        all_points[row[2]] += 1

for team, games in n_games_team.items():
    print('{:>9}: {} {} {} {} {}'.format(team, games, n_win_team[team], n_draw_team[team], n_lose_team[team], all_points[team]))
```

Класс Фамилия Рост

Класс обозначается только числом. Буквенные модификаторы не используются. Номер класса может быть от 1 до 11 включительно. В фамилии нет пробелов, а в качестве роста используется натуральное число, но при подсчёте среднего требуется вычислить значение в виде вещественного числа.

Выведите таблицу размером n × n, заполненную числами от 1 до n^2 по спирали, выходящей из левого верхнего угла и закрученной по часовой стрелке, как показано в примере (здесь n=5):

Sample Input:

5

Sample Output:

1  2  3  4  5

16 17 18 19 6

15 24 25 20 7

14 23 22 21 8

13 12 11 10 9

```python
n = int(input())            # для примера n = 5
m = [[0 for i in range(n)] for j in range(n)]

min_y = 1       # минимальное значение координаты по вертикали
min_x = 0       # минимальное значение координаты по горизонтали
max_y = n-1     # макс по вертикали
max_x = n-1     # макс по горизонтали

y = 0           # координаты по вертикали
x = -1          # координаты по горизонтали
count = 1
while True:
    while x < max_x:        # пока координата по горизонтали меньше своего максимального значения
        x += 1              # происходит заполнение матрицы слева направо:
        m[y][x] = count     # Изначально - m[0][0] = 1, затем m[0][1] = 2... (строка зафиксирована)
        count += 1
    max_x -= 1           # сужение границы по горизонтали (максимальное значение по горизонтали уменьшается)

    while y < max_y:     # заполнение матрицы сверху вниз
        y += 1
        m[y][x] = count  # фиксированный столбец
        count += 1
    max_y -= 1           # сужение границы по вертикали (максимальное значение по вертикали уменьшается)

    while x > min_x:     # заполнение матрицы справа налево:
        x -= 1
        m[y][x] = count  
        count += 1
    min_x += 1           # сужение границы по горизонтали (минимальное значение по горизонтали увеличивается)

    while y > min_y:     # заполнение матрицы снизу вверх:
        y -= 1
        m[y][x] = count
        count += 1
    min_y += 1           # сужение границы по вертикали (увеличивая минимальное значение по вертикали)

    if x == (n - 1) // 2 and y == n // 2:
        break

for i in m:
    for j in i:
        print(j, end=' ')
    print()
```

Задача: Необходимо вычислить n-е число Фибоначчи (напомним, что F0 = 0, F1 = 1, Fn = Fn-2 + Fn-1 , при n >= 2)

```python
cache = {}
def fib(n):
    assert n >= 0
    if n not in cache:
        cache[n] = n if n <= 1 else fib(n - 1) + fib(n - 2)
    return cache[n]

n = int(input())
print(fib(n))
```
